# Distribution Plots in Seaborn

```python
import seaborn as sns

#  view matplotlib in notebook
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,hist_kws=dict(edgecolor='black'))
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30,hist_kws=dict(edgecolor='black'))
```

#  Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

#  Pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

#  Rug Plot

```python
sns.rugplot(tips['total_bill'])
```
