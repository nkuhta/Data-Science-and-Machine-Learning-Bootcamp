```python
import matplotlib.pyplot as plt
```

```python
#  See inside jupyter notebook
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y = x**2
y
```

##  Functional Plotting method

```python
plt.plot(x,y)
plt.show()  # required if you're not in jupyter notebook
```

## Labeling Axis and Title 

```python
#  Axis label
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#plt.show()
```

##  Multiple Plots 

```python
plt.subplot(1,2,1)  #  (rows,columns,graph)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

##  Object Oriented API Method
* More Control and efficient

```python
# figure object
fig = plt.figure()

#  add_axes(left,bottom,width,height)
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()
#  add_axes(left,bottom,width,height)
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.07,0.42,0.5,0.5])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title("smaller plot")
axes1.set_title('large plot')
```
