```python
import matplotlib.pyplot as plt
```

```python
#  See inside jupyter notebook
%matplotlib inline
```

```python
import numpy as np
```

```python
#  x = 11 values from 0 to 5(inclusive)
x=np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

##  Functional Plotting method

```python
plt.plot(x,y)
plt.show()  # required if you're not in jupyter notebook
```

## Labeling Axis and Title 

```python
#  Axis label
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#plt.show()
```

##  Multiple Plots 

```python
plt.subplot(1,2,1)  #  (rows,columns,graph)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

#  Object Oriented API Method
* More Control and efficient

```python
# figure object
fig = plt.figure()

#  add_axes(left,bottom,width,height)
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

## multiple embedded axes

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()
#  add_axes(left,bottom,width,height)
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.07,0.42,0.5,0.5])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title("smaller plot")
axes1.set_title('large plot')
```

## subplots()

```python
#  recall the data
np.array([x,y])
```

```python
(fig,axes) = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
#plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
(fig,axes) = plt.subplots(nrows=1,ncols=2)
#  use axes array objects
axes[0].plot(x,y)
```

```python
(fig,axes) = plt.subplots(nrows=1,ncols=2)
#  use axes array objects
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
(fig,axes) = plt.subplots(nrows=1,ncols=2)
#  use axes array objects
axes[0].plot(x,y)
axes[1].plot(y,x)

axes[0].set_title('First Plot')
axes[1].set_title('Second Plot')

#  for anything overlapping
plt.tight_layout()
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2))  # 3 inches by 2 inches

ax = fig.add_axes([0,0,1.2,1.2])
ax.plot(x,y)
```

```python
(fig,axes) = plt.subplots(nrows=2,ncols=1,figsize=(3,5))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## Saving a Figure
---

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label="X Squared")
ax.plot(x,x**3,label='X Cubed')

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')

ax.legend()
```

##  savefig

```python
#  Save figure and set dpi and bbox = tight to fit axes
fig.savefig('my_picture.jpg',dpi=200,bbox_inches='tight')  #png,jpg,pdf etc.  
```

#  Colors, Linewidth, Alpha

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,alpha=0.2)   #  Also RGB Hex code custom colors
```

#  lw, linestyle (ls)

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=3,linestyle='--')
plt.show()
```

# Markers

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=1,ls='-',marker='o',markersize=6,
       markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='red')
plt.show()
```

#  Control of Axes appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')
#  Axes limits 
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

#  Scatter Plot

```python
plt.scatter(x,y)
plt.show()
```

# Histogram

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
```

# Box Plot 

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);   
```
