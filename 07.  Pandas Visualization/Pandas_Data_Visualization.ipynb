#  Built-in Data Visualization with Pandas

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
fig1=df1['A'].hist(bins=30,linewidth=1,edgecolor='white',alpha=1)
fig1.grid(linestyle='--',color='green',alpha=0.3)
```

##  Histogram

```python
df1['A'].plot.hist()
```

#  Plot Types

## Area Plots

```python
df2.plot.area(alpha=0.4)
```

```python
df2
```

## Bar Plot

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## Line Plot

```python
df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

## Scatter plots

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

## Box Plots

```python
df2.plot.box()
```

# Bivariate Data, hexbin

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```
